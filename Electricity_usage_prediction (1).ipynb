**Handling Null Values**

```python

# Drop rows where Sub_metering_3 is null
df = df.dropna(subset=['Sub_metering_3'])
df
```

```python
# Fill missing values in Sub_metering_3 with the median and assign it back
df['Sub_metering_3'] = df['Sub_metering_3'].fillna(df['Sub_metering_3'].median())
```

```python
# Display the updated DataFrame (for the first few rows)
print("Updated DataFrame:")
print(df.head())
```

```python
# Specify the columns to exclude (Date and Time columns)
columns_to_exclude = ['Date', 'Time']

# Drop the Date and Time columns and fill missing values in the remaining columns
df_filtered = df.drop(columns=columns_to_exclude).fillna(0)

# Count the null values after filling
null_values = df_filtered.isnull().sum()
print("\nNull Values Count After Filling (Excluding Date and Time Columns):")
print(null_values)
```
